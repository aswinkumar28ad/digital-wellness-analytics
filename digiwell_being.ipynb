```python
# Group by stress level and find the average social media hours for each level
stress_screen_time = df.groupby('stress_level')['daily_social_media_hours'].mean().reset_index()
print(stress_screen_time)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Group the data by physical activity and calculate the average social media hours
# We use .round(2) to keep the numbers clean and easy to read
avg_data = df.groupby('physical_activity')['daily_social_media_hours'].mean().reset_index()

print("--- 📊 ACTUAL CALCULATED AVERAGES ---")
print(avg_data.round(2).to_string(index=False))
print("\n------------------------------------")

# 2. Set up our bar chart visual
plt.figure(figsize=(10, 6))

sns.barplot(
    x='physical_activity',
    y='daily_social_media_hours',
    data=avg_data,
    palette='viridis'  # A beautiful, clean color gradient
)

# 3. Add clear titles and labels for your portfolio
plt.title('Average Daily Social Media Hours by Physical Activity Baseline', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Physical Activity Level', fontsize=12, labelpad=10)
plt.ylabel('Average Social Media Usage (Hours)', fontsize=12, labelpad=10)

# Add value labels on top of each bar so we can see the exact numbers on the chart
for index, row in avg_data.iterrows():
    plt.text(index, row.daily_social_media_hours + 0.1, f"{row.daily_social_media_hours:.2f}",
             color='black', ha="center", fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Group by sleep quality and find the average screen time before sleep
sleep_trend = df.groupby('sleep_quality')['screen_time_before_sleep'].mean().reset_index()

# Sort it logically from Poor to Good sleep so the chart reads nicely
sleep_trend['sleep_quality'] = pd.Categorical(sleep_trend['sleep_quality'], categories=['Poor', 'Fair', 'Good'], ordered=True)
sleep_trend = sleep_trend.sort_values('sleep_quality').reset_index(drop=True)

print("--- 📊 AVERAGES FOR SLEEP QUALITY ---")
print(sleep_trend.round(2).to_string(index=False))
print("\n------------------------------------")

# 2. Draw the Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(
    x='sleep_quality',
    y='screen_time_before_sleep',
    data=sleep_trend,
    palette='Reds_r' # Uses a red gradient (dark red to light red)
)

plt.title('Average Screen Time Before Sleep by Sleep Quality Level', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Reported Sleep Quality', fontsize=12)
plt.ylabel('Average Screen Time Before Sleep (Hours)', fontsize=12)

# Add exact value labels on top of the bars
for index, row in sleep_trend.iterrows():
    plt.text(index, row.screen_time_before_sleep + 0.05, f"{row.screen_time_before_sleep:.2f} hrs",
             color='black', ha="center", fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Group by social interaction level and calculate average stress level
stress_trend = df.groupby('social_interaction_level')['stress_level'].mean().reset_index()

# Sort the categories logically from low to high social interaction
stress_trend['social_interaction_level'] = pd.Categorical(
    stress_trend['social_interaction_level'],
    categories=['low', 'medium', 'high'],
    ordered=True
)
stress_trend = stress_trend.sort_values('social_interaction_level').reset_index(drop=True)

print("--- 📊 AVERAGES FOR STRESS LEVEL ---")
print(stress_trend.round(2).to_string(index=False))
print("\n-----------------------------------")

# 2. Draw the Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(
    x='social_interaction_level',
    y='stress_level',
    data=stress_trend,
    palette='Purples' # Clean purple gradient
)

plt.title('Average Stress Level by Social Interaction Tier', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Social Interaction Level', fontsize=12)
plt.ylabel('Average Stress Level (Scale 1-10)', fontsize=12)

# Add exact value labels on top of the bars
for index, row in stress_trend.iterrows():
    plt.text(index, row.stress_level + 0.1, f"{row.stress_level:.2f}",
             color='black', ha="center", fontsize=10, fontweight='bold')

plt.ylim(0, 10) # Set limit to match the 1-10 stress scale bounds
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create clean age groups so the rows aren't cluttered
# Slicing the data into distinct lifestyle/student cohorts
def age_bucket(age):
    if age <= 15: return 'Early Teens (13-15)'
    elif age <= 18: return 'Late Teens (16-18)'
    else: return 'Young Adults (19+)'

df['age_group'] = df['age'].apply(age_bucket)

# 2. Pivot the data: Rows = Age Groups, Columns = Platforms, Values = Avg Academics
pivot_table = df.pivot_table(
    index='age_group',
    columns='platform_usage',
    values='academic_performance',
    aggfunc='mean'
)

# 3. Plot the clean, color-coded grid
plt.figure(figsize=(11, 5))
sns.heatmap(
    pivot_table,
    annot=True,          # Shows the exact score inside the box
    fmt=".2f",           # Rounds the scores to 2 decimal places
    cmap='RdYlGn',       # Red (Lower Grades) to Green (Higher Grades) gradient
    linewidths=1,
    cbar_kws={'label': 'Average Academic Performance'}
)

plt.title('Academic Performance Intersection Matrix: Age vs. Platform Used', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Social Media Platform Used', fontsize=12, labelpad=10)
plt.ylabel('Age Cohort', fontsize=12, labelpad=10)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Round social media hours to the nearest whole hour so we can group them cleanly
df['social_media_hours_rounded'] = df['daily_social_media_hours'].round()

# 2. Calculate the average stress and anxiety for each hour mark
trend_data = df.groupby('social_media_hours_rounded')[['stress_level', 'anxiety_level']].mean().reset_index()

# 3. Plot the dual trend lines
plt.figure(figsize=(12, 6))

# Plot Stress Line
sns.lineplot(
    x='social_media_hours_rounded', y='stress_level', data=trend_data,
    marker='o', color='crimson', linewidth=3, label='Average Stress Level (1-10)'
)

# Plot Anxiety Line
sns.lineplot(
    x='social_media_hours_rounded', y='anxiety_level', data=trend_data,
    marker='s', color='darkorange', linewidth=3, label='Average Anxiety Level (1-10)'
)

plt.title('The Digital Tipping Point: Screen Duration vs. Mental Health Risk Accrual', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Daily Social Media Exposure (Rounded to Nearest Hour)', fontsize=12, labelpad=10)
plt.ylabel('Psychological Risk Score (Scale 1-10)', fontsize=12, labelpad=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)

plt.tight_layout()
plt.show()
```
